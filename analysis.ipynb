```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ファイルからデータを読み込む
open_data = pd.read_csv('/Users/yasudayuuya/class/DSprogramming/DSfinal/open.csv')
local_data = pd.read_csv('/Users/yasudayuuya/class/DSprogramming/DSfinal/local.csv')

open_data['time'] = pd.to_datetime(open_data['time'])
local_data['time'] = pd.to_datetime(local_data['time'])

plt.figure(figsize=(10, 6))

plt.plot(open_data['time'], open_data['temperature'], label='Open Temperature')
plt.plot(local_data['time'], local_data['temperature'], label='Local Temperature')

plt.title('Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')

plt.legend()

# グラフを表示
plt.show()
```

このグラフからわかる通り、気象庁の気温のデータと比較したときに日当たりの良い時間帯は、かなり高い気温を示すことがわかった。
